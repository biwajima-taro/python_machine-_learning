```python
import numpy as np
np.random.seed(0)

N_ARMS=4

class Env:
    thetas=[0.1,0.1,0.2,0.3]
    @classmethod
    def react(cls,arm):
        return 1 if np.random.random()<cls.thetas[arm] else 0
    @classmethod
    def opt(cls):
        """
        return the index of max cell
        """
        return np.argmax(cls.thetas)

    

class EpsillonGreedyAgent:
    def __init__(self,epsilon=0.1):
        self.epsilon=epsilon
        self.counts=np.zeros(N_ARMS)
        self.values=np.zeros(N_ARMS)
    def get_arm(self):
        if np.random.random()<self.epsilon:
            arm=np.random.randint(N_ARMS)
        else:
            arm=np.argmax(self.values)
        return arm
    def sample(self,arm,reward):
        self.counts[arm]+=1
        self.values[arm]=((self.counts[arm]-1)*self.values[arm]+reward)/self.counts[arm]

        
Env.opt()        
```

```python
from matplotlib import pyplot as plt

def sim(Agent,N=1000,T=1000,**kwargs):
    selected_arms=[[0 for  _ in range(T)] for _ in range(T)]
    earned_rewards=[[0 for _ in range(T)] for _ in range(T)]
    
    for n in range(N):
        agent=Agent(**kwargs)
        for t in range(T):
            arm=agent.get_arm()
            reward=Env.react(arm)
            agent.sample(arm,reward)
            selected_arms[n][t]=arm
            earned_rewards[n][t]=reward
    return np.array(selected_arms),np.array(earned_rewards)
            
arms_eg,rewards_eg=sim(EpsillonGreedyAgent)
print(arms_eg==Env.opt())
acc=np.mean(arms_eg==Env.opt(),axis=0)

print(acc)
```
